##**Demandas de ML**

El proceso de Análisis Exploratorio de Datos (EDA) consiste en comprender cómo están estructurados los datos que queremos analizar.

Es un proceso de naturaleza investigativa, donde intentamos entender diversas características, como: los valores presentes en las columnas, los tipos de estructura de datos, verificar si son datos cualitativos o cuantitativos, si hay valores faltantes o inusuales.

Por lo tanto, en este momento, las preguntas sobre los datos siempre son bienvenidas. Estas guiarán todo el proceso de análisis, y a través de las herramientas disponibles, como nuestro querido Pandas, buscaremos respuestas.

Algunas preguntas que podemos hacer en este momento son:

¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

##**Remover registros inconsistentes**

Algunos registros en la base de datos no tienen mucho sentido, por ejemplo:

Departamentos con un valor de alquiler igual a 0.

Departamentos con un valor de condominio igual a 0.

Estos registros son inconsistentes, por lo tanto, debemos eliminarlos de nuestra base de datos.

```python
df.query('Valor == 0 | Condominio == 0')
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
df_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
df.drop(df_remover, axis=0, inplace=True)
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
df.head()
```

```python
df.Tipo.unique()
```

```python
## Como todos los departementos se quitara la columna Tipo

df.drop('Tipo', axis=1, inplace=True)
df.sample(10)
```

##**Aplicar filtros solicitados**

En este momento, el equipo de ML desea evaluar solo propiedades en escenarios específicos, como:

Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

Con base en esto, necesitamos aplicar los filtros necesarios para obtener esta información específica.

### 1. Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200

```python
seleccion1 = df['Habitaciones'] == 1
```

```python
seleccion1
```

```python
type(seleccion1)
```

```python
df[seleccion1]
```

```python
seleccion2 = df['Valor'] < 4200
df[seleccion2]
```

```python
df[seleccion2].max()
```

```python
# filtro1 = df[seleccion1 & seleccion2]

filtro1 = (seleccion1) & (seleccion2)
df[filtro1]
```

```python
df1 = df[filtro1]
```

### 2. Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

```python
filtro2 = (df['Habitaciones'] >= 2) & (df['Valor'] < 10500) & (df['Area'] > 70)
df[filtro2]
```

```python
df2 = df[filtro2]
```
