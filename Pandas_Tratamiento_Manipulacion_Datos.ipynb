### *Números en strings*

```python
datos['precio']
```

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

### *Transformación en múltiples columnas*

```python
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
#No funciona asi:
datos[['cuota_deposito', 'cuota_limpieza']].apply(lambda x: x.replace('$', '').replace(',', ''))
#ya que aply funciona solo cuando es con 1 unica columna
```

```python
#Cuando es una unica columna es una seria, pero si hay mas ya es un DataFrame
#Para DataFrame se usa applymap
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

#**3. Datos de texto**


---

```python
datos.head()
```

```python
datos['descripcion_local']
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos['descripcion_local'][3169]
```

### *Eliminando caracteres con Regex*

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex=True)
#^: Significa negación, que se reemplazara lo contrario a: | a-z | A-Z | 0-9 | \- | \' |
# letras de la a-z en minuscula, de la A-Z en mayuscula, numeros del 0 al 9. Signifique que le digo a python que reemplace todo lo que hay en el texto menos eso
#\-: que no sustituya los guinones
#\': que no reemplace las comillas simples
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)', ' ', regex=True)
# Que busque los guiones - pero que no tengan una palabra o un caracter (?<!\w) a la izquierda y tampoco a la derecha (?!\w)
```

```python
datos['descripcion_local'].head()
```

### *Tokenizacion de strings*

```python
datos['descripcion_local'].str.split()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
datos['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
datos['comodidades'] = datos['comodidades'].str.split(',')
datos.head()
```
