### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total por cada tienda sumando la columna 'Precio'
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
# Mostrar resultados
print(f"Ingreso total Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
# Ventas por categoría para Tienda 2
ventas_categoria_t2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 2:\n", ventas_categoria_t2, "\n")
```

```python
# Ventas por categoría para Tienda 3
ventas_categoria_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 3:\n", ventas_categoria_t3, "\n")
```

```python
# Ventas por categoría para Tienda 4
ventas_categoria_t4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 4:\n", ventas_categoria_t4, "\n")
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio por tienda
calificacion_media_t1 = tienda['Calificación'].mean()
calificacion_media_t2 = tienda2['Calificación'].mean()
calificacion_media_t3 = tienda3['Calificación'].mean()
calificacion_media_t4 = tienda4['Calificación'].mean()
```

```python
# Mostrar resultados
print(f"Valoración media Tienda 1: {calificacion_media_t1:.2f}")
print(f"Valoración media Tienda 2: {calificacion_media_t2:.2f}")
print(f"Valoración media Tienda 3: {calificacion_media_t3:.2f}")
print(f"Valoración media Tienda 4: {calificacion_media_t4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
```

```python
def mostrar_ventas_extremas_por_tienda(df, nombre_tienda):
    # Agrupamos y contamos las ventas por producto
    ventas_productos = df.groupby('Producto').size().sort_values(ascending=False)

    # Mostramos el más vendido y el menos vendido
    producto_mas_vendido = ventas_productos.idxmax()
    cantidad_mas_vendido = ventas_productos.max()

    producto_menos_vendido = ventas_productos.idxmin()
    cantidad_menos_vendido = ventas_productos.min()

    print(f"{nombre_tienda} - Producto más vendido: {producto_mas_vendido} ({cantidad_mas_vendido} ventas)")
    print(f"{nombre_tienda} - Producto menos vendido: {producto_menos_vendido} ({cantidad_menos_vendido} ventas)\n")

    # Visualización
    top_5 = ventas_productos.head(5)
    bottom_5 = ventas_productos.tail(5)

    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    top_5.plot(kind='bar', ax=ax[0], title=f"{nombre_tienda} - Top 5 productos más vendidos", color='green')
    bottom_5.plot(kind='bar', ax=ax[1], title=f"{nombre_tienda} - Top 5 productos menos vendidos", color='red')
    plt.tight_layout()
    plt.show()
```

```python
# Aplicar a cada tienda
mostrar_ventas_extremas_por_tienda(tienda, "Tienda 1")
mostrar_ventas_extremas_por_tienda(tienda2, "Tienda 2")
mostrar_ventas_extremas_por_tienda(tienda3, "Tienda 3")
mostrar_ventas_extremas_por_tienda(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
envio_promedio_t1 = tienda['Costo de envío'].mean()
envio_promedio_t2 = tienda2['Costo de envío'].mean()
envio_promedio_t3 = tienda3['Costo de envío'].mean()
envio_promedio_t4 = tienda4['Costo de envío'].mean()
```

```python
# Mostrar resultados
print(f"Costo de envío promedio - Tienda 1: ${envio_promedio_t1:,.2f}")
print(f"Costo de envío promedio - Tienda 2: ${envio_promedio_t2:,.2f}")
print(f"Costo de envío promedio - Tienda 3: ${envio_promedio_t3:,.2f}")
print(f"Costo de envío promedio - Tienda 4: ${envio_promedio_t4:,.2f}")
```

# 6. Graficos

1. Gráfico de barras: Ingresos totales por tienda

```python
# Ingresos por tienda
ingresos = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos ($)')
plt.xlabel('Tiendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2. Gráfico de pastel: Distribución de categorías - Tienda 1

```python
categorias_t1 = tienda['Categoría del Producto'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(categorias_t1, labels=categorias_t1.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Categorías - Tienda 1')
plt.axis('equal')
plt.show()
```

3. Gráfico de dispersión: Relación entre precio del producto y costo de envío - Tienda 1


```python
plt.figure(figsize=(8,5))
plt.scatter(tienda['Precio'], tienda['Costo de envío'], alpha=0.6, color='purple')
plt.title('Precio vs. Costo de Envío - Tienda 1')
plt.xlabel('Precio del Producto ($)')
plt.ylabel('Costo de Envío ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
